## Étape 2:

Au lieu d'utiliser PostMan, faisons une requête via une API.

```python
import requests

flower_example = {
"sepal_length":5.1,
"sepal_width":3.5,
"petal_length":1.4,
"petal_width":0.2
}
```

```python
# à faire en local sur Jupyter Notebook (ou autre IDE)
r = requests.post("http://localhost:5000/api/flower",json=flower_example)
```

```python
if r.status_code == 200:
    print(f"Success: {r.text}")
else:
    print(f"Failure: {r.text}")
```

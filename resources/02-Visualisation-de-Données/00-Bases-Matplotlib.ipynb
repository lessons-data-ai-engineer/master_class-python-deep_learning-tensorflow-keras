## Style

L'une des principales raisons d'apprendre les bases complètes est de voir comment les interactions de style affectent l'API.

```
plt.plot(x,y)
```

```
plt.plot(x,y)
plt.title('Titre')
plt.xlabel('Label X')
plt.ylabel('Label Y');
```

```
plt.plot(x,y)

# Axes et marques
plt.xlim(0,2)
plt.ylim(100,300)


# Etiquetage 
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label');
```

```
plt.plot(x,y,color='red')

# Axes et marques
plt.xlim(0,2)
plt.ylim(100,300)


# Etiquetage
plt.title('Titre')
plt.xlabel('Label X')
plt.ylabel('Label Y');
```

```
plt.plot(x,y,color='red',marker='o',markersize=20,linestyle='--')

# Axes et marques
plt.xlim(0,2)
plt.ylim(100,300)


# Etiquetage
plt.title('Titre')
plt.xlabel('Label X')
plt.ylabel('Label Y');
```

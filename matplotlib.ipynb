```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
x = [0,1,2]
y = [100,200,300]

plt.plot(x,y)
plt.show()
```

```python
housing = pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3],
                        'price':[100,120,190,200,230,310,330,305]})
```

```python
plt.scatter(housing['rooms'], housing['price'])
plt.show()
```

```python
plt.plot(x,y, color='red', marker='o', markersize=15, linestyle='--')

plt.xlim(0,2)
plt.ylim(0,300)

plt.title('Titre')
plt.xlabel('Label x')
plt.ylabel('Label y')
plt.show()
```
